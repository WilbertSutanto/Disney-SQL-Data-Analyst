```python
url = 'https://www.consumeraffairs.com/entertainment/disney-plus.html?page='
```

```python
url
```

```python
for pagenumber in range(1,3):
    business_request = requests.get(url + str(pagenumber))
    soup = bs(business_request.text, 'html.parser')
    reviews = soup.findAll('div', attrs={'class':'rvw js-rvw'})
    
    
    for review in reviews:
#print(review)
    
    # user 
    # rvw-aut__inf-nm 
        user = review.find('strong', attrs={'class':'rvw-aut__inf-nm'})
        if user is not None:
        #print(user.text)
        #print(type(user.text))
    
            user_match = re.search('^[a-zA-Z]+',user.text)
            if user_match is not None:
                print('user_name:', user_match.group(0))
        
        
            state_match = re.search('[a-zA-Z]+$', user.text )
            if state_match is not None:
                print('user_state:', state_match.group(0))
        
        
    
    # review date
    # ca-txt-cpt 
        review_date = review.find('span', attrs={'class':'ca-txt-cpt'})
        if review_date is not None:
            print('review_date:', review_date.text.replace('Original review: ',''))
        
    

        review_text = review.find('p')
        if review_text is not None:
            print('review_text:', review_text.text)
        

        review_rating = review.find('meta', attrs={'itemprop':'ratingValue'})
        if review_rating is not None:
            print('rating:', review_rating['content'])
    
        print('-'*100)
    
```

```python
reviews_data = {
    'user_name': [],
    'user_state': [],
    'review_date' : [],
    'review_text' : [],
    'rating' : []
}


for pagenumber in range(1,3):
    business_request = requests.get(url + str(pagenumber))
    soup = bs(business_request.text, 'html.parser')
    reviews = soup.findAll('div', attrs={'class':'rvw js-rvw'})
    
    
    for review in reviews:

        user = review.find('strong', attrs={'class':'rvw-aut__inf-nm'})
        if user is not None:
      
    
            user_match = re.search('^[a-zA-Z]+',user.text)
            if user_match is not None:
                print('user_name:', user_match.group(0))
                reviews_data['user_name'].append(user_match.group(0)) 
            else:
                print('user_name:', '')
                reviews_data['user_name'].append('')
                
                
        
            state_match = re.search('[a-zA-Z]+$', user.text )
            if state_match is not None:
                print('user_state:', state_match.group(0))
                reviews_data['user_state'].append(state_match.group(0))
            else:
                print('user_state:','')
                reviews_data['user_state'].append('')
                
        
    
   
        review_date = review.find('span', attrs={'class':'ca-txt-cpt'})
        if review_date is not None:
            print('review_date:', review_date.text.replace('Original review: ',''))
            reviews_data['review_date'].append(review_date.text.replace('Original review: ',''))
        else:
            print('review_date:', '')
            reviews_data['review_date'].append('')
        
    

        review_text = review.find('p')
        if review_text is not None:
            print('review_text:', review_text.text)
            reviews_data['review_text'].append(review_text.text)
        else:
            print('review_text:', '')
            reviews_data['review_text'].append('')
            
        

        review_rating = review.find('meta', attrs={'itemprop':'ratingValue'})
        if review_rating is not None:
            print('rating:', review_rating['content'])
            reviews_data['rating'].append(review_rating['content'])
        else:
            print('rating:', '')
            reviews_data['rating'].append('')
            
    
        print('-'*100)
```

```python
reviews_df = pd.DataFrame(reviews_data)
```
